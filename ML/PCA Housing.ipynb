## **PCA**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import fetch_california_housing
```

```python
data= fetch_california_housing()
```

```python
df= pd.DataFrame(data.data, columns= data.feature_names)
df['MedVal']= data.target
```

```python
df.info()
```

```python
df.describe()
```

```python
df.hist(figsize=(10,8), bins=20, edgecolor='k')
plt.tight_layout()
plt.show()
```

```python
X= df.drop('MedVal', axis=1)
y= df['MedVal']
```

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
```

```python
scalar_data= scaler.fit_transform(X)
scalar_data
```

```python
from sklearn.decomposition import PCA

pca= PCA(n_components=2)
pca
```

```python
pca_data= pca.fit_transform(scalar_data)
pca_data
```

```python
pca.explained_variance_
```

```python
plt.scatter(pca_data[:, 0], pca_data[:, 1], marker='*', edgecolors='k', s=300)
plt.xlabel('1st PC')
plt.ylabel('2nd PC')
plt.show()
```
